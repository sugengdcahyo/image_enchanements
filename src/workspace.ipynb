```python
def count_elements(seq) -> dict:
    hist = {}
    idx=0
    for x in range(seq.shape[0]):
        for i in seq[x]:
            hist[i] = hist.get(i, 0) + 1
    return hist
```

```python
r_count = count_elements(r)
g_count = count_elements(g)
b_count = count_elements(b)
```

```python
def sorted_by_keys(seq):
    dict_items = seq.items()
    sorted_items = sorted(dict_items)
    return dict(sorted_items)
```

```python
r_sorted = sorted_by_keys(r_count)
g_sorted = sorted_by_keys(g_count)
b_sorted = sorted_by_keys(b_count)
```

```python
plt.bar(list(r_sorted.keys()), r_sorted.values(), color='r')
plt.rcParams['figure.dpi'] = 100
plt.show()
```

```python
plt.bar(list(g_sorted.keys()), g_sorted.values(), color='g')
plt.rcParams['figure.dpi'] = 100
plt.show()
```

```python
plt.bar(list(b_sorted.keys()), b_sorted.values(), color='b')
plt.rcParams['figure.dpi'] = 100
plt.show()
```
